```python
#importing releted libraries

import selenium
import pandas as pd 
from selenium import webdriver
import warnings 
warnings.filterwarnings("ignore")
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
#connecting driver

driver=webdriver.Chrome(r"chromedriver.exe")
driver
```

```python
driver.get("https://www.flipkart.com/")
```

```python
start=1
end=100
Review_of_laptops=[]
for page in range(start,end):
    review=driver.find_elements(By.XPATH,'//p[@class="_2-N8zT"]')
    for i in review:
        Review_of_laptops.append(i.text)
    next_button=driver.find_elements(By.XPATH,'//a[@class="_1LKTO3"]')
    
    time.sleep(3)
    
Review_of_laptops
```

```python
start=1
end=111
Ratings_of_laptops=[]
for page in range(start,end):
    rating=driver.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]')[0:990]
    for i in rating:
        Ratings_of_laptops.append(i.text)
    next_button=driver.find_elements(By.XPATH,'//a[@class="_1LKTO3"]')
    
    time.sleep(3)
    
Ratings_of_laptops
```

```python
print(len(Review_of_laptops))
print(len(Ratings_of_laptops))
```

```python
df=pd.DataFrame({"Review":Review_of_laptops, "Rating":Ratings_of_laptops})
df
```
